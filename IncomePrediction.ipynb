# Diviser en variables catégoriques et numériques

```python
cat_data=[]
num_data=[]
for i, c in enumerate(data.dtypes):
    if c==object:
        cat_data.append(data.iloc[:,i])
    else:
        num_data.append(data.iloc[:,i])
cat_data=pd.DataFrame(cat_data).transpose()
num_data=pd.DataFrame(num_data).transpose()
```

```python
cat_data
```

```python
for i in cat_data:
    cat_data[i]=le.fit_transform(cat_data[i])
cat_data
```

```python
X=pd.concat([cat_data,num_data],axis=1)
```

```python
X
```

```python
for col in X:
    plt.figure()
    sns.heatmap(pd.crosstab(data['income_>50K'], data[col]), annot=True, fmt='d')
```

# Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(I, Y, test_size=0.3, random_state=40)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pgrid={"splitter":["best","random"],
             "max_depth":range(2,20,1),
             "min_samples_leaf":range(1,15,1),
             "min_samples_split":range(2,20,1) 
            }
```

```python
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid=pgrid, cv=5)
grid_search.fit(X_train, y_train)
grid_search.best_estimator_.score(X_test, y_test)
```

```python
grid_search.best_estimator_
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(12,12))
tree.plot_tree(grid_search.best_estimator_,rounded=True,filled=True)
plt.show()
```
